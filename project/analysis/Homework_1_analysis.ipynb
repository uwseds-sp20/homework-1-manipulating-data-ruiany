```python
import pandas as pd
import numpy as np
from datetime import datetime
from dateutil.parser import parse
import matplotlib.pyplot as plt
```

```python
# Read in csv as dataframe

df_bike = pd.read_csv('/Users/ruianyang/Documents/GradSchool/MSDS/DATA514/homework-1-manipulating-data-ruiany/project/data/Fremont_Bridge_Bicycle_Counter.csv')
```

```python
# Add hour of the day column

df_bike['Hour'] = pd.DatetimeIndex(pd.to_datetime(df_bike['Date'],format="%m/%d/%Y %I:%M:%S %p")).hour
```

```python
# Add year column

df_bike['Year'] = pd.DatetimeIndex(pd.to_datetime(df_bike['Date'],format="%m/%d/%Y %I:%M:%S %p")).year
```

```python
# Create a dataframe with the subset of data from the year 2016

df_subset_2016 = df_bike[df_bike['Year'] >= 2016]
```

```python
# Plot the total counts by hour

plt.figure(figsize = (9,6))
plt.plot(df_bike.Hour, df_bike['Fremont Bridge Total'])
plt.xticks(np.arange(0, 24, step=1))
plt.xlabel('Hour of the Day',fontsize = 15)
plt.ylabel('Number of Bikes', fontsize = 15)
plt.title('Fremont Bridge Bike Counts', fontsize = 20)
plt.show()
```

```python
# Plot the sum total counts by hour

plt.figure(figsize = (9,6))
plt.plot(list(range(24)), df_bike[['Fremont Bridge Total', 'Hour']].groupby('Hour').agg('sum')['Fremont Bridge Total'].tolist())
plt.xticks(np.arange(0, 24, step=1))
plt.xlabel('Hour of the Day',fontsize = 15)
plt.ylabel('Sum Number of Bikes', fontsize = 15)
plt.title('Fremont Bridge Bike Counts', fontsize = 20)
plt.show()
```

```python
# Calculating the busiest hour of the day

groups = df_bike[['Fremont Bridge Total', 'Hour']].groupby('Hour').agg('sum')
print(groups[groups['Fremont Bridge Total'] == groups['Fremont Bridge Total'].max()])
```
